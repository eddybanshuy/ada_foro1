Uso de EDA y fuentes de datos para un caso particular

Dataset Utilizado es de Kaggle " https://www.kaggle.com/datasets/faresashraf1001/supermarket-sales"

1. Cargar Dataset

```python
# Librerías principales
import pandas as pd
import numpy as np

# Cargar dataset (ejemplo: archivo CSV descargado de Kaggle)
df = pd.read_csv("data/supermarket_sales.csv")

# Mostrar primeras filas
print(df.head())
```

2. Información del Dataset

```python
# Nombre de columnas
print("Columnas:", df.columns.tolist())

# Tamaño del dataset
print("Filas y columnas:", df.shape)

# Descripción estadística
print(df.describe())

# Tipos de datos
print(df.info())
```

```python
# Visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")
```

3. Tareas de EDA
🔹 Univariantes

```python
# Histograma de ventas totales
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Distribución de Ventas Totales")
plt.show()

# Conteo de clientes por género
sns.countplot(x="Gender", data=df)
plt.title("Distribución de Clientes por Género")
plt.show()
```

🔹 Multivariantes

```python
# Relación entre método de pago y ventas
plt.figure(figsize=(8,5))
sns.boxplot(x="Payment", y="Sales", data=df)
plt.title("Método de Pago vs Ventas")
plt.show()

# Correlación entre variables numéricas
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

4. Preparación de Datos

```python
# Manejo de valores nulos
print(df.isnull().sum())

# Conversión de fechas
df['Date'] = pd.to_datetime(df['Date'])

# Crear columna adicional: mes de la compra
df['Month'] = df['Date'].dt.month
df[['Date','Month']].head()
```
